```julia
using PyPlot

using LinearAlgebra
using SparseArrays
using Arpack

using GSL
using DifferentialEquations
using Sundials

using PyCall
using LsqFit

using Printf
```

```julia
## Rate Equations Model 
function rate_solve(du,u,p,t)
    Γ41 = p[1];
    Γ24 = p[2];
    Γ34 = p[3];
    Γ12 = p[4];
    Γ13 = p[5];
    
    du[1] = -Γ41*u[1]*(1-u[4]) + Γ12*u[2]*(1-u[1]) + Γ13*u[3]*(1-u[1]);
    du[2] = Γ24*u[4]*(1-u[2]) - Γ12*u[2]*(1-u[1]);
    du[3] = Γ34*u[4]*(1-u[3]) - Γ13*u[3]*(1-u[1]);
    du[4] = Γ41*u[1]*(1-u[4]) - Γ24*u[4]*(1-u[2]) - Γ34*u[4]*(1-u[3]);
end;
```

```julia
# Fix ratio value based on experimental data -- low-current approximation
# expt_ratio = y_unpat[1][argmin(y_unpat[3])]
expt_ratio = 2.5
Γ24_0 = 10.0
Γ34_0 = Γ24_0*expt_ratio
Γ12_0 = Γ24_0/10.
Γ13_0 = Γ12_0/20.
pump_vec = [0.001:0.001:1.5;];
n_pumps = size(pump_vec,1);
pops = zeros(n_pumps, 4);
emission_ratio = zeros(n_pumps);

for ii = 1 : n_pumps
    p = [pump_vec[ii], Γ24_0, Γ34_0, Γ12_0, Γ13_0];
    u0 = zeros(4);
    u0[1] = 1.0;

    tspan = (0.0, 1e2);
    prob = ODEProblem(rate_solve, u0, tspan, p)
#     sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);
    sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
    pops[ii, :] = sol.u[size(sol.t,1)];
    emission_ratio[ii] = (p[5]*sol.u[size(sol.t,1)][3]) / (p[4]*sol.u[size(sol.t,1)][2]);    
end
```

```julia
function model(x, p, sol_flag = false)
    ratio1 = p[1]
    ratio4 = p[2]
    scaling = p[3]
    
    Γ24_0 = 10.
    Γ34_0 = Γ24_0*ratio4    
    Γ12_0 = Γ24_0/10.
    Γ13_0 = Γ12_0/ratio1
        
    emission_ratio = zeros(size(x,1))
    t_vec = 0.
    u_vec = 0.

    for ii = 1 : size(x,1);
        params = [scaling*x[ii], Γ24_0, Γ34_0, Γ12_0, Γ13_0];

        u0 = zeros(4);
        u0[1] = 1.0;

        tspan = (0.0, 5e2);
        prob = ODEProblem(rate_solve, u0, tspan, params)
        sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);        
#         sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
#         sol = solve(prob, CVODE_BDF(), reltol=1e-6, abstol=1e-6);                

        emission_ratio[ii] = (params[5]*sol.u[size(sol.t,1)][3]) / (params[4]*sol.u[size(sol.t,1)][2]);    
        t_vec = sol.t
        u_vec = sol.u
    end
    if sol_flag == false
        return emission_ratio
    else
        return emission_ratio, t_vec, u_vec
    end
end;
```

```julia
function model2(x, p, sol_flag = false)
    ratio1 = p[1]
    ratio4 = p[2]
    scaling = p[3]
    
    Γ24_0 = 10.
    Γ34_0 = Γ24_0*ratio4    
    Γ12_0 = Γ24_0/10.
    Γ13_0 = Γ12_0/ratio1
        
    emission_ratio = zeros(size(x,1))
    t_vec = 0.
    u_vec = 0.

    for ii = 1 : size(x,1);
        params = [scaling*x[ii], Γ24_0, Γ34_0, Γ12_0, Γ13_0];

        u0 = zeros(4);
        u0[1] = 1.0;

        tspan = (0.0, 5e2);
        prob = ODEProblem(rate_solve, u0, tspan, params)
#         sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);        
#         sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
        sol = solve(prob, CVODE_BDF(), reltol=1e-6, abstol=1e-6);        

        emission_ratio[ii] = (params[5]*sol.u[size(sol.t,1)][3]) / (params[4]*sol.u[size(sol.t,1)][2]);    
        t_vec = sol.t
        u_vec = sol.u
    end
    if sol_flag == false
        return emission_ratio
    else
        return emission_ratio, t_vec, u_vec
    end
end;
```

# Comparison of different methods (unpatterned)

```julia
function evaluate_method(fit, title)
    p1, p2, p3 = fit.param
    margin_of_error = margin_error(fit)
    confidence_intervals = confidence_interval(fit)
    std_error = stderror(fit)
    print(title, " \n")    
    @printf("Unpat. all: Γ24 = 10., Γ34 = Γ24*%.2e, Γ12 = Γ24 / 10. and Γ13 = Γ12 / %.2e Γ14 = %.2e * current (μA) \n", p2, p1, p3)
    @printf("Unpat. fit parameters: %.2e, %.2e, %.2e \n", p1, p2, p3)
    print("Unpat. error: \n")
    map(x -> @printf("%.2e  ",x), std_error)
    print("\n Unpat. error (%): \n")    
    map(x -> @printf("%.2e  ",x), std_error./fit.param*100)    
    print("\n \n")        
end;
evaluate_method(fit_bounds_unpat, "Experimental uncertainty")
evaluate_method(fit_bounds_unpat00, "Remove low current plateau")
evaluate_method(fit_bounds_unpat02, "Numerical uncertainty from first fit")
```

```julia
function fit_comparison(fit_unpat, fit_pat)
    p1_thick4, p2_thick4, p3_thick4 = fit_pat.param
    p1_unpat, p2_unpat, p3_unpat = fit_unpat.param
    enhancement = p1_thick4/p1_unpat
    current_scaling_ratio = p3_thick4/p3_unpat
    @printf("Patterned to unpatterned Γ12 / Γ13 enhancement = %.2e \n", enhancement)
    @printf("Patterned to unpatterned current scaling ratio = %.2e", current_scaling_ratio)        
    print("\n \n")            
end;
```

```julia
# Analysis of the fits:
fit_unpat = fit_bounds_unpat02
evaluate_method(fit_bounds_thick4_00, "Experimental uncertainty, unclamped (Tsit5)")
fit_comparison(fit_unpat, fit_bounds_thick4_00)

evaluate_method(fit_bounds_thick4_01, "Experimental uncertainty, clamped (Tsit5 and CVODE)")
fit_comparison(fit_unpat, fit_bounds_thick4_01)

evaluate_method(fit_bounds_thick4_10, "Numerical uncertainty, unclamped (Tsit5)")
fit_comparison(fit_unpat, fit_bounds_thick4_10)

evaluate_method(fit_bounds_thick4_11, "Numerical uncertainty, clamped (Tsit5 and CVODE)")
fit_comparison(fit_unpat, fit_bounds_thick4_11)
```

```julia
# Fix ratio value based on experimental data -- low-current approximation
# expt_ratio = y_unpat[1][argmin(y_unpat[3])]
expt_ratio = 2.5
Γ24_0 = 100.0
Γ34_0 = Γ24_0*expt_ratio
Γ12_0 = Γ24_0/100.
Γ13_0 = Γ12_0/20.
pump_vec = [0.001:0.001:1.5;];
n_pumps = size(pump_vec,1);
pops = zeros(n_pumps, 4);
emission_ratio = zeros(n_pumps);

for ii = 1 : n_pumps
    p = [pump_vec[ii], Γ24_0, Γ34_0, Γ12_0, Γ13_0];
    u0 = zeros(4);
    u0[1] = 1.0;

    tspan = (0.0, 1e3);
    prob = ODEProblem(rate_solve, u0, tspan, p)
#     sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);
    sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
    pops[ii, :] = sol.u[size(sol.t,1)];
    emission_ratio[ii] = (p[5]*sol.u[size(sol.t,1)][3]) / (p[4]*sol.u[size(sol.t,1)][2]);    
end
```

# Method 1: Use experimental uncertainty as weight on LSE

```julia
function model(x, p, sol_flag = false)
    ratio1 = p[1]
    ratio4 = p[2]
    scaling = p[3]
    
    Γ24_0 = 10.
    Γ34_0 = Γ24_0*ratio4    
    Γ12_0 = Γ24_0/100.
    Γ13_0 = Γ12_0/ratio1
        
    emission_ratio = zeros(size(x,1))
    t_vec = 0.
    u_vec = 0.

    for ii = 1 : size(x,1);
        params = [scaling*x[ii], Γ24_0, Γ34_0, Γ12_0, Γ13_0];

        u0 = zeros(4);
        u0[1] = 1.0;

        tspan = (0.0, 5e3);
        prob = ODEProblem(rate_solve, u0, tspan, params)
        sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);        
#         sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
#         sol = solve(prob, CVODE_BDF(), reltol=1e-6, abstol=1e-6);                

        emission_ratio[ii] = (params[5]*sol.u[size(sol.t,1)][3]) / (params[4]*sol.u[size(sol.t,1)][2]);    
        t_vec = sol.t
        u_vec = sol.u
    end
    if sol_flag == false
        return emission_ratio
    else
        return emission_ratio, t_vec, u_vec
    end
end;
```

```julia
## Unpatterned, use experimental uncertainty for weights 
# Roughly insensitive to initial state (within 5-10% )
xdata_unpat = y_unpat[3]*1e6
ydata_unpat = y_unpat[1]*1.
lb = [0., 0., 0.]
ub = [100., 100.,0.01]

p0_bounds = [100., 2.5, 1/200.]

wt = 1 ./ y_unpat[2]

fit_bounds_unpat = curve_fit(model, xdata_unpat, ydata_unpat, wt, p0_bounds, lower=lb, upper=ub)
# fit_bounds_unpat = curve_fit(model, xdata_unpat, ydata_unpat, p0_bounds)
```

```julia
# plot(xdata_unpat, model(xdata_unpat, p0_bounds), "o")
xdata_unpat_plot = [minimum(xdata_unpat):1.0:maximum(xdata_unpat);]
plot(sort(xdata_unpat_plot), model(sort(xdata_unpat_plot), fit_bounds_unpat.param), "--", label = "model")
errorbar(xdata_unpat, ydata_unpat, yerr = y_unpat[2], fmt = "o", label = "data (unpat.)");
legend()
xlabel("Current (μA)")
ylabel("Green to red ratio")
title("Unpatterned fit")
savefig("unpatterned_fit.svg")
```
