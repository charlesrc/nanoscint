```julia
function ene_loss_ratio(ene_kev)
    res = zeros((length(ene_kev),))
    for i = 1:size(ene_kev,1);
        res[i] = ene_loss[argmin(abs.(ene_kev[i].-envals_kev))]
    end;
    return res
end;
```

```julia
## Rate Equations Model 
function rate_solve(du,u,p,t)
    13 = p[1]; # Green to Excited
    31 = p[2]; # Excited to Green
    23 = p[3]; # Red to Excited
    32 = p[4]; # Excited to Red 
    
    du[1] = -13*u[1]*(1-u[3]) + 31*u[3]*(1-u[1]);
    du[2] = -23*u[2]*(1-u[3]) + 32*u[3]*(1-u[2]);
    du[3] = 13*u[1]*(1-u[3]) - 31*u[3]*(1-u[1]) + 23*u[2]*(1-u[3]) - 32*u[3]*(1-u[2]);
end;
```

# Full data fit (voltage and current dependence)
- Create model function that takes into account current density 
- Fit with two different techniques: with and without low-current plateau
- Compare uncertainty, etc. (remove data points for unpatterned at low-current, to make sure uncertainty is not bogus)
- TODO next: include depth dependence (overlap between EELD and silica layer)

```julia
function model(x, p, sol_flag = false)
    # x is power (in mW)
    ratio1 = p[1]
    scaling = p[2]
    
    24_0 = 10.
    34_0 = 24_0*ratio4    
    12_0 = 24_0/10.
    13_0 = 12_0/ratio1
        
    emission_ratio = zeros(size(x,1))
    t_vec = 0.
    u_vec = 0.

    for ii = 1 : size(x,1);
        params = [scaling*x[ii], 24_0, 34_0, 12_0, 13_0];

        u0 = zeros(4);
        u0[1] = 1.0;

        tspan = (0.0, 5e2);
        prob = ODEProblem(rate_solve, u0, tspan, params)
#         sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);        
#         sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
        sol = solve(prob, CVODE_BDF(), reltol=1e-6, abstol=1e-6);                

        emission_ratio[ii] = (params[5]*sol.u[size(sol.t,1)][3]) / (params[4]*sol.u[size(sol.t,1)][2]);    
        t_vec = sol.t
        u_vec = sol.u
    end
    if sol_flag == false
        return emission_ratio
    else
        return emission_ratio, t_vec, u_vec
    end
end;
```

```julia
function model2(x, p, sol_flag = false)
    ratio1 = p[1]
    ratio4 = p[2]
    scaling = p[3]
    
    24_0 = 10.
    34_0 = 24_0*ratio4    
    12_0 = 24_0/10.
    13_0 = 12_0/ratio1
        
    emission_ratio = zeros(size(x,1))
    t_vec = 0.
    u_vec = 0.

    for ii = 1 : size(x,1);
        params = [scaling*x[ii], 24_0, 34_0, 12_0, 13_0];

        u0 = zeros(4);
        u0[1] = 1.0;

        tspan = (0.0, 5e2);
        prob = ODEProblem(rate_solve, u0, tspan, params)
#         sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);        
#         sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
        sol = solve(prob, CVODE_BDF(), reltol=1e-6, abstol=1e-6);        

        emission_ratio[ii] = (params[5]*sol.u[size(sol.t,1)][3]) / (params[4]*sol.u[size(sol.t,1)][2]);    
        t_vec = sol.t
        u_vec = sol.u
    end
    if sol_flag == false
        return emission_ratio
    else
        return emission_ratio, t_vec, u_vec
    end
end;
```

```julia
## Unpatterned, use experimental uncertainty for weights 
# Roughly insensitive to initial state (within 5-10% )
lb = [0., 0., 0.]
ub = [50., 100.,0.1]

p0_bounds = [25., 2.5, 1/20.]

wt = 1 ./ yerr_unpat

fit_bounds_unpat = curve_fit(model, x_unpat, y_unpat, wt, p0_bounds, lower=lb, upper=ub)
```

```julia
# plot(xdata_unpat, model(xdata_unpat, p0_bounds), "o")
x_unpat_plot = [minimum(x_unpat):1.0:maximum(x_unpat);]
plot(sort(x_unpat_plot), model(sort(x_unpat_plot), fit_bounds_unpat.param), "--", label = "model", color = "blue")
errorbar(x_unpat, y_unpat, yerr = yerr_unpat, fmt = "o", label = "data (unpat.)", color = "blue");
legend()
xlabel("Beam power (mW)")
ylabel("Green to red ratio")
title("Unpatterned fit (volt. and curr.)")
```

```julia
# Check time convergence of solver 
x_test = x_unpat[1]
_, t_test, u_test = model(x_test, fit_bounds_unpat.param, true)

# x_test = xdata_thick4[end]
# _, t_test, u_test = model(x_test, fit_bounds_thick4.param, true)
plot(t_test, u_test);
```

# Comparison of different methods (unpatterned)

```julia
function evaluate_method(fit, title)
    p1, p2, p3 = fit.param
    margin_of_error = margin_error(fit)
    confidence_intervals = confidence_interval(fit)
    std_error = stderror(fit)
    print(title, " \n")    
    @printf("24 = 10., 34 = 24*%.2e, 12 = 24 / 10. and 13 = 12 / %.2e 14 = %.2e * current (A) \n", p2, p1, p3)
    @printf("fit parameters: %.2e, %.2e, %.2e \n", p1, p2, p3)
    print("error: \n")
    map(x -> @printf("%.2e  ",x), std_error)
    print("\n error (%): \n")    
    map(x -> @printf("%.2e  ",x), std_error./fit.param*100)    
    print("\n \n")        
end;
evaluate_method(fit_bounds_unpat, "Experimental uncertainty")
evaluate_method(fit_bounds_unpat00, "Remove low current plateau")
```

```julia
function fit_comparison(fit_unpat, fit_pat)
    p1_thick4, p2_thick4, p3_thick4 = fit_pat.param
    p1_unpat, p2_unpat, p3_unpat = fit_unpat.param
    enhancement = p1_thick4/p1_unpat
    current_scaling_ratio = p3_thick4/p3_unpat
    @printf("Patterned to unpatterned 12 / 13 enhancement = %.2e \n", enhancement)
    @printf("Patterned to unpatterned current scaling ratio = %.2e", current_scaling_ratio)        
    print("\n \n")            
end;
```
