```julia
using PyPlot

using LinearAlgebra
using SparseArrays
using Arpack

using GSL
using QuantumOptics
using DifferentialEquations
```

```julia
function rate_solve(du,u,p,t)
    Γ41 = p[1];
    Γ24 = p[2];
    Γ34 = p[3];
    Γ12 = p[4];
    Γ13 = p[5];
    
    du[1] = -Γ41*u[1]*(1-u[4]) + Γ12*u[2]*(1-u[1]) + Γ13*u[3]*(1-u[1]);
    du[2] = Γ24*u[4]*(1-u[2]) - Γ12*u[2]*(1-u[1]);
    du[3] = Γ34*u[4]*(1-u[3]) - Γ13*u[3]*(1-u[1]);
    du[4] = Γ41*u[1]*(1-u[4]) - Γ24*u[4]*(1-u[2]) - Γ34*u[4]*(1-u[3]);
end
```

```julia
p = [0.1,1.0,2.0,0.01,0.01];
X = zeros(4);
X[1] = 1.0;
u0 = X;

tspan = (0.0,1e3);
prob = ODEProblem(rate_solve,u0,tspan,p)
sol = solve(prob,Tsit5(),reltol=1e-6, abstol=1e-6);
```

```julia
plot(sol.u)
```

```julia
sol.u[size(sol.t,1)]
```

```julia
pump_vals = [1e-5:0.01:0.5;];
n_pumps = size(pump_vals,1);

decay_ratios = [1:0.1:1.5;];
n_ratios = size(decay_ratios,1);

pop_ratio = zeros(n_ratios,n_pumps);
pops = zeros(n_ratios,n_pumps,4);
emission_ratio = zeros(n_ratios,n_pumps);

for ii=1:n_ratios
    for jj=1:n_pumps
        p = [pump_vals[jj],1.0,1.0*decay_ratios[ii],0.01,0.01];
        X = zeros(4);
        X[1] = 1.0;
        u0 = X;

        tspan = (0.0,1e3);
        prob = ODEProblem(rate_solve,u0,tspan,p)
        sol = solve(prob,Tsit5(),reltol=1e-6, abstol=1e-6);
        pop_ratio[ii,jj] = sol.u[size(sol.t,1)][3] / sol.u[size(sol.t,1)][2];
        pops[ii,jj,:] = sol.u[size(sol.t,1)];
        emission_ratio[ii,jj] = (p[5]*sol.u[size(sol.t,1)][3]) / (p[4]*sol.u[size(sol.t,1)][2]);
        
    end
end
```

```julia
for ii=1:n_ratios
    plot(pump_vals, emission_ratio[ii,:])
end
plot(pump_vals,ones(n_pumps),linestyle="--",color="k")
xlabel("Pump strength p[1]")
ylabel("Emission ratio (green/red)")
```

```julia
for ii=1:n_ratios
    plot(pump_vals, pops[ii,:,1])
end
xlabel("Pump strength p[1]")
ylabel("Emission ratio (green/red)")
```

```julia
pump_vals = [0.01:0.01:1;];
n_pumps = size(pump_vals,1);

decay_ratios = [1:0.5:50;];
n_ratios = size(decay_ratios,1);

pop_ratio = zeros(n_ratios,n_pumps);
emission_ratio = zeros(n_ratios,n_pumps);

for ii=1:n_ratios
    for jj=1:n_pumps
        p = [pump_vals[jj],1.0,1.0*decay_ratios[ii],0.02,0.01];
        X = zeros(4);
        X[1] = 1.0;
        u0 = X;

        tspan = (0.0,1e3);
        prob = ODEProblem(rate_solve,u0,tspan,p)
        sol = solve(prob,Tsit5(),reltol=1e-6, abstol=1e-6);
        pop_ratio[ii,jj] = sol.u[size(sol.t,1)][3] / sol.u[size(sol.t,1)][2];
        emission_ratio[ii,jj] = (p[5]*sol.u[size(sol.t,1)][3]*(1-sol.u[size(sol.t,1)][1])) /
        (p[4]*sol.u[size(sol.t,1)][2]*(1-sol.u[size(sol.t,1)][1]));
    end
end
```

```julia
pcolormesh(pump_vals,decay_ratios,pop_ratio)
xlabel("Pump strength p[1]")
ylabel("Decay ratio p[3]/p[2]")
colorbar()
```

```julia
pcolormesh(pump_vals,decay_ratios, 1 ./ emission_ratio)
xlabel("Pump strength p[1]")
ylabel("Decay ratio p[3]/p[2]")
colorbar()
```

```julia
for ii=1:n_ratios
    plot(pump_vals, 1 ./ emission_ratio[ii,:])
end
plot(pump_vals,ones(n_pumps),linestyle="--",color="k")
xlabel("Pump strength p[1]")
ylabel("Emission ratio (red/green)")
```
