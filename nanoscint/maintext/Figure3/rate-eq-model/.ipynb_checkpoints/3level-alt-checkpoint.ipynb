```julia
function ene_loss_ratio(ene_kev)
    res = zeros((length(ene_kev),))
    for i = 1:size(ene_kev,1);
        res[i] = ene_loss[argmin(abs.(ene_kev[i].-envals_kev))]
    end;
    return res
end;
```

# Three-level model

```julia
## Rate Equations Model 
function rate_solve(du,u,p,t)
    13 = p[1]; # Green to Excited
    31 = p[2]; # Excited to Green
    23 = p[3]; # Red to Excited
    32 = p[4]; # Excited to Red 
    
    du[1] = -13*u[1]*(1-u[3]) + 31*u[3]*(1-u[1]);
    du[2] = -23*u[2]*(1-u[3]) + 32*u[3]*(1-u[2]);
    du[3] = 13*u[1]*(1-u[3]) - 31*u[3]*(1-u[1]) + 23*u[2]*(1-u[3]) - 32*u[3]*(1-u[2]);
end;
```

```julia
function rate_solve_nopointer(u,p)
    du = ones((3,1))
    13 = p[1]; # Green to Excited
    31 = p[2]; # Excited to Green
    23 = p[3]; # Red to Excited
    32 = p[4]; # Excited to Red 
    
    du[1] = -13*u[1]*(1-u[3]) + 31*u[3]*(1-u[1]);
    du[2] = -23*u[2]*(1-u[3]) + 32*u[3]*(1-u[2]);
    du[3] = 13*u[1]*(1-u[3]) - 31*u[3]*(1-u[1]) + 23*u[2]*(1-u[3]) - 32*u[3]*(1-u[2]);
    return du 
end;
```

```julia
pump_vals = [1e-5:0.01:2;];
n_pumps = size(pump_vals, 1);

decay_ratios = [0.1:0.2:2.0;];
n_ratios = size(decay_ratios, 1);

pop_ratio = zeros(n_ratios, n_pumps);
pops = zeros(n_ratios, n_pumps, 3);
emission_ratio = zeros(n_ratios, n_pumps);

for ii=1:n_ratios
    for jj=1:n_pumps
        p = [pump_vals[jj], decay_ratios[ii], pump_vals[jj], 1.0];
        X = zeros(3);
        X[1] = 1.0;
        X[2] = 1.0;
#         X[3] = 1.0;        
        u0 = X;

        tspan = (0.0,1e2);
        prob = ODEProblem(rate_solve, u0, tspan, p)
        sol = solve(prob, Tsit5(), reltol=1e-6, abstol=1e-6);
        pop_ratio[ii,jj] = sol.u[size(sol.t,1)][1] / sol.u[size(sol.t,1)][2];
        pops[ii,jj,:] = sol.u[size(sol.t,1)];
        emission_ratio[ii,jj] = (p[2]*(1-sol.u[size(sol.t,1)][1])) / (p[4]*(1-sol.u[size(sol.t,1)][2]));
        
    end
end
```

```julia
for ii=1:n_ratios
    plot(pump_vals, emission_ratio[ii,:])
end
plot(pump_vals,ones(n_pumps),linestyle="--",color="k")
xlabel("Pump strength p[1]")
ylabel("Emission ratio (green/red)")
```

```julia
p = [10.0,0.5,2.0,1.0];
X = zeros(3);
X[1] = 1.0;
X[2] = 1.0;
# X[3] = 1.0;
u0 = X;

tspan = (0.0,1e2);
prob = ODEProblem(rate_solve,u0,tspan,p)
sol = solve(prob,CVODE_BDF(),reltol=1e-8, abstol=1e-8);
# sol = solve(prob, Tsit5(), reltol=1e-10, abstol=1e-10);
plot(sol.u);

@printf("steady state from ODE solve: %.2e, %.2e, %.2e \n", sol.u[size(sol.t,1)][1], sol.u[size(sol.t,1)][2], sol.u[size(sol.t,1)][3])
## Function for NL Solver 
function f!(F,u)
    norm(rate_solve(F, u, p, 0))
end;
@printf("CV check: %.2e \n", norm(rate_solve_nopointer(sol.u[size(sol.t,1)], p)))
nlsolve(f!, sol.u[size(sol.t, 1)], ftol = 1e-20, method = :newton)
# mcpsolve(f!, [0.,0.,0.], [1.0,1.0,1.0], rand(3,1), reformulation = :smooth, autodiff = :forward)
```

# Full data fit (voltage and current dependence)

```julia
function model(x, p, sol_flag = false)
    # x is power (in mW)
    ratio1 = p[1] # green to red
    scaling = p[2] # pump scaling
    scaling_ratio = p[3] # pump ratio (green to red)
    
    32_0 = 1.
    31_0 = 32_0*ratio1 
        
    emission_ratio = zeros(size(x,1))
    t_vec = 0.
    u_vec = 0.

    for ii = 1 : size(x,1);
        params = [scaling_ratio*scaling*x[ii], 31_0, scaling*x[ii], 32_0];

        u0 = zeros(3);
        u0[1] = 1.0;

        tspan = (0.0, 1e1);
        prob = ODEProblem(rate_solve, u0, tspan, params)

#         sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8);        
#         sol = solve(prob, Rosenbrock23(), reltol=1e-6, abstol=1e-6);
        sol = solve(prob, CVODE_BDF(), reltol=1e-6, abstol=1e-6);                

        emission_ratio[ii] = (params[2]*(1-sol.u[size(sol.t,1)][1])) / (params[4]*(1-sol.u[size(sol.t,1)][2]));
        
        t_vec = sol.t
        u_vec = sol.u
    end
    if sol_flag == false
        return emission_ratio
    else
        return emission_ratio, t_vec, u_vec
    end
end;
```

```julia
## Unpatterned, use experimental uncertainty for weights 
# Roughly insensitive to initial state (within 5-10% )
# lb = [0.1, 0.00001, 0.1]
# ub = [5., 0.1,5.0]
# p0_bounds = [1.5, 1/100.,3.0]

lb = [0.1, 0.00062854, 47.3301]
ub = [10., 0.00062854, 47.3301]
p0_bounds = [0.7, 0.00062854, 47.3301]


wt = 1 ./ yerr_unpat

fit_bounds_unpat = curve_fit(model, x_unpat, y_unpat, wt, p0_bounds, lower=lb, upper=ub)
# fit_bounds_unpat = curve_fit(model, x_unpat, y_unpat, p0_bounds, lower=lb, upper=ub)
```

```julia
function evaluate_method(fit, x, title)
    p1, p2, p3 = fit.param
    margin_of_error = margin_error(fit)
    confidence_intervals = confidence_interval(fit)
    std_error = stderror(fit)
    rates_min = [p2*p3*minimum(x),p1,p2*minimum(x),1.]
    rates_max = [p2*p3*maximum(x),p1,p2*maximum(x),1.]
    print(title, " \n")    
    @printf("32 = 1., 31 = 32*%.2e, 13 = %.2e * deposited energy, 23 = %.2e * 13 \n", p1, p2, p3)
    @printf("fit parameters: %.2e, %.2e, %.2e \n", p1, p2, p3)
    print("error: \n")
    map(x -> @printf("%.2e  ",x), std_error)
    print("\n error (%): \n")    
    map(x -> @printf("%.2e  ",x), std_error./fit.param*100)    
    print("\n Min. rates : \n ")    
    print("[ 13 , 31, 23, 32] \n [")
    map(x -> @printf("%.2e  ",x), rates_min)
    print("] \n  Max. rates : \n")    
    print("[ 13 , 31, 23, 32] \n [")
    map(x -> @printf("%.2e  ",x), rates_max)    
    print("] \n \n")        
end;
evaluate_method(fit_bounds_unpat, x_unpat, "Experimental uncertainty")
```

```julia
function fit_comparison(fit_unpat, fit_pat)
    p1_thick4, p2_thick4, p3_thick4 = fit_pat.param
    p1_unpat, p2_unpat, p3_unpat = fit_unpat.param
    enhancement = p1_thick4/p1_unpat
    current_scaling_ratio = p2_thick4/p2_unpat
    current_scaling_ratio_ratio = p3_thick4/p3_unpat
    @printf("Patterned to unpatterned 31 / 32 enhancement = %.2e \n", enhancement)
    @printf("Patterned to unpatterned current scaling ratio = %.2e \n", current_scaling_ratio)        
    @printf("Patterned to unpatterned  13 / 12 enhancement = %.2e \n", current_scaling_ratio_ratio)        
    print("\n \n")            
end;
```

```julia
# Analysis of the fits:
fit_unpat = fit_bounds_unpat
evaluate_method(fit_bounds_thick4_00, x_thick4, "Experimental uncertainty, unclamped (Tsit5)")
fit_comparison(fit_unpat, fit_bounds_thick4_00)
```
