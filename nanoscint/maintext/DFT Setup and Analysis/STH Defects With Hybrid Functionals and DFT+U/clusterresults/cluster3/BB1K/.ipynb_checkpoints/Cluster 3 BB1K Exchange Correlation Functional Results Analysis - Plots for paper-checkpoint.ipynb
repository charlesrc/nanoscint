```julia
id = "31"
excitations = np.loadtxt("cluster$(id).Excitations", skiprows=6, usecols=[3, 6]);
rexcitations = zeros(500)
broadening = .2
for trans in eachrow(excitations)
    diffe, strength = trans
    rexcitations += 2/3*strength*exp.(-(diffe*27.2 .- range(1, 3, length=500) ).^2 ./broadening^2 )
end
```

```julia
id = "31"
spinup = np.fromfile("cluster$(id).eigenvals")[1:28];
spindn = np.fromfile("cluster$(id).eigenvals")[29:end];
# μ = parse(Float64, split(readlines("cluster$(id).eigStats")[contains.(readlines("cluster$(id).eigStats"), "mu")][1])[3])
μ = parse(Float64, split(readlines("cluster$(id).eigStats")[occursin.("mu", readlines("cluster$(id).eigStats"))][1])[3])
```

```julia
plot(eV*reshape(np.repeat(spinup, 10), (10, 28) )[:, 11:16], linewidth=3,  color=:blue)
#hline!([μ*27.2], linewidth=5, color=:red)
#ylims!(-25, 0)

#b = plot(eV*reshape(np.repeat(spindn, 10), (10, 28) ), linewidth=3, legend=false, color=:blue)
#hline!([μ*27.2], linewidth=5, color=:red)
#ylims!(-25, 0)
#plot(a, b, xticks=false)
```

```julia
momenta_up = np.reshape(np.fromfile("cluster$(id).momenta", dtype=np.complex), (2, 3, 28, 28))[1, :, :, : ];
momenta_dn = np.reshape(np.fromfile("cluster$(id).momenta", dtype=np.complex), (2, 3, 28, 28))[2, :, :, : ];
momenta_upsquared = ((abs.(momenta_up[1, :, : ])).^2+ (abs.(momenta_up[2, :, : ])).^2+ (abs.(momenta_up[3, :, : ])).^2);
momenta_dnsquared = ((abs.(momenta_dn[1, :, : ])).^2+ (abs.(momenta_dn[2, :, : ])).^2+ (abs.(momenta_dn[3, :, : ])).^2);
```

```julia
transitions = []
energies = []
rightspin = spinup
for (i, e1) in enumerate(rightspin)
    println("i=$i")
    e1 < μ || continue
    for (j, e2) in enumerate(rightspin)
            println("i=$i")

        e2 > μ || continue
        ((e2-e1) < 5/27.2) || continue
        rightspin == spinup ? push!(transitions, (momenta_upsquared[i, j]/(e2-e1) )) : push!(transitions, (momenta_dnsquared[i, j]/(e2-e1) ))
        push!(energies, (e2-e1)*27.2)
    end
end
```

```julia
energiesgauss = zeros(500)
broadening=0.1
for (i, (transition, energy)) in enumerate(zip(transitions, energies))
    energiesgauss += 2/3*transition*exp.( -((energy .- range(1, 3, length=500)).^2/broadening^2))
end
excitations = np.loadtxt("cluster$(id).Excitations", skiprows=6, usecols=[3, 6]);
rexcitations = zeros(500)
for trans in eachrow(excitations)
    diffe, strength = trans
    rexcitations += 2/3*strength*diffe*exp.(-(diffe*27.2 .- range(1, 3, length=500) ).^2 ./broadening^2 )
end
```

```julia
plot(1240 ./ range(1, 3, length=500), energiesgauss / maximum(energiesgauss), linewidth=5, label="Momentum Matrix Elements")
plot( 1240 ./ range(1, 3, length=500), rexcitations / maximum(rexcitations), linewidth=5, label="Dipole Matrix Elements")
#vline!((1240 ./ range(1, 3, length=500))[argmaxima(energiesgauss)], linewidth=3, linestyle=:dash, label="" )
#xlims!(400, 800)
#xlabel!("λ(nm)")
#yticks!(Float64[])
#title!("Cluster 3 Spectrum", size=(500, 400))
#savefig()
```

The figure above corresponds to silica at a bond angle very close (5 degrees from) the relaxed bond angle. This indicates that the transitions that we're seeing are very close to corresponding to actual silica and not a horribly disfigured defect.
