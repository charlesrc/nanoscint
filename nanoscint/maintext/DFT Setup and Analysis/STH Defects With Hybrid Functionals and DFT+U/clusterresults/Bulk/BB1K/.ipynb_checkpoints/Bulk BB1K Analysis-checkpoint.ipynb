```julia
excitations = np.loadtxt("alpha.Excitations", skiprows=6, usecols=[3, 6]);
eV=27.2
```

```julia
broadenedXdipole = zeros(500)
broadening=0.1
energies = range(1.3, 5, length=500)
for (e, rsq) in zip(excitations[:, 1], excitations[:, 2])
    broadenedXdipole += rsq*e*exp.(-((energies .- 27.2*e).^2)/broadening^2)
end

plot(energies, broadenedXdipole, linewidth=5, color=:red, legend=false, size=(600, 400), title="Bulk Excitation Spectrum")
xlims!(1, 3)

ylims!(0, 0.4)
```

```julia
plot!(1240 ./ energies, broadenedXdipole)
```

```julia
momenta = np.reshape(np.fromfile("alpha.momenta", np.complex), (2*8, 3, 36, 36))
```

```julia
eups = np.fromfile( "alpha.eigenvals")[1:36*8]*27.2
edns = np.fromfile( "alpha.eigenvals")[36*8+1:36*16]*27.2
a = plot(transpose(reshape(repeat(eups, 10), (36*8, 10))), legend=false, ylims=[-5,10 ] , linewidth=3, color=:blue)
b = plot(transpose(reshape(repeat(edns, 10), (36*8, 10))), legend=false, ylims=[-5,10] , xticks=nothing,size=(400, 600), linewidth=3, color=:red)

plot(a, b)
```

```julia
broadenedX = zeros(500)
broadening=0.1
energies = range(1.3, 5, length=500)
for kpoint in 0:7
    for band1 in 1:36
        for band2 in 1:36
            e1 = edns[kpoint*36+band1]; e2 = edns[kpoint*36+band2]
            e1 < 0 || continue
            e2 > 0 || continue
            #println(e1, "  ", e2)
            e = e2-e1
            psquared = (abs(momenta[kpoint+4+1, 1, band1, band2]))^2+(abs(momenta[kpoint+4+1, 2, band1, band2]))^2+(abs(momenta[kpoint+4+1, 3, band1, band2]))^2
            broadenedX += psquared/e*exp.(-((energies .- e).^2)/broadening^2)
        end
    end
end

plot(energies, broadenedX/maximum(broadenedX), linewidth=5, color=:red, legend=false, size=(600, 400), title="Bulk Excitation Spectrum (Momentum Matrix Elements)")
plot!(energies, broadenedXdipole/maximum(broadenedXdipole), linewidth=5, color=:black, legend=false, size=(600, 400), title="Bulk Excitation Spectrum (Momentum Matrix Elements)")

xlims!(1.3, 3)
yticks!(Float64[])

xlabel!("Energy (eV)")
```

```julia
#plot(1240 ./ energies, broadenedX/maximum(broadenedX), linewidth=5, color=:red, legend=false, size=(600, 400), title="Bulk Excitation Spectrum (Momentum Matrix Elements)")
plot(1240 ./ energies, broadenedXdipole/maximum(broadenedXdipole), linewidth=5, color=:black, legend=false, size=(600, 400), title="Bulk Excitation Spectrum")

xlims!(450, 800)
yticks!(Float64[])
xlabel!(" (nm)", xtickfontsize=12, xguidefontsize=20)
```
