```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.io import savemat, loadmat
```

```python
file_mat = loadmat('res/thinfilm_sweep.mat')
Veff_array = file_mat['Veff_array']
eps_CsI_array = file_mat['eps_CsI_array']
wl_vec = file_mat['wl_vec']
rad_vec = file_mat['rad_vec']
etch_depth_vec = file_mat['etch_depth_vec']
```

```python
plt.imshow(np.squeeze(Veff_array), aspect=0.1, extent = [np.min(wl_vec), np.max(wl_vec), np.min(rad_vec), np.max(rad_vec)])
plt.show()
```

```python
np.trapz(np.multiply(Veff_array, np.imag(eps_CsI_array)), wl_vec)
```

```python
plt.plot(np.squeeze(wl_vec), np.squeeze(np.multiply(Veff_array, np.imag(eps_CsI_array))))
```

```python
plt.plot(np.squeeze(wl_vec), np.imag(eps_CsI_array[0,1,:]))
```
